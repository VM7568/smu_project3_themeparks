```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import requests
import json
```

```python
data = pd.read_csv('safer-parks-accident-dataset.csv', encoding='cp1252')
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#Make an address
df['Address'] = df.acc_city + ', ' + df.acc_state
df.head()
```

```python
df['year'].value_counts()
```

```python
df2 = df.loc[df['acc_city'] != '0']
df2.info()
```

```python
sample_df = df2.sample(n=1000, random_state=1234)
sample_df.info()
```

```python
sample_df.head(25)
```

```python
df2 = df.loc[df['year'] >= 2000]
df2['year'].value_counts()
```

```python
sample_df = df2.sample(n=1000, random_state=1)
sample_df.info()
```

```python
sample_df.to_csv('sample_df.csv')
```

```python
sample_df.to_json("data.json", orient="columns")
```

```python
dfgpavg=sample_df.groupby(["year", "device_category"]).mean().round(0)
cols=['num_injured']
dfgpavg2=dfgpavg.loc[:,cols]
dfgpavg2.head(20)
```

```python
sns.relplot(data=dfgpavg, x="num_injured", y="device_category", hue="year")
```

```python
dfgpavg.to_json("data_group_average.json", orient="columns")
```

```python
dfgp_count=sample_df.groupby(["year", "device_category"])["num_injured"].sum()
dfgp_count.head(50)
```

```python
dfgp_count.info()
```

```python
dfgp_count.to_json("data_group_num_injured.json", orient="columns")
```

```python
dfgp2=sample_df.groupby(["year", "device_category"]).sum()
cols=['num_injured']
dfgp3=dfgp2.loc[:,cols]
dfgp3.reset_index(inplace=True)
dfgp3.head(5)
```

```python
dfgp3.to_csv('dfgp4.csv')
```

```python
total_injured=sample_df.groupby(["year"])["num_injured"].sum()
total_injured
```
