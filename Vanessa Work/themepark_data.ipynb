```python
#Get location data from opwnweathermap.org

url = f"http://api.openweathermap.org/geo/1.0/direct?q={address}, USA&appid={API_KEY}"
response = requests.get(url)

data = response.json()

if (len(data) > 0):
    location = data[0]
    print(location)
```

```python
#Loop to get latitude & longitude all locations

rows = []

for address in city_df.address:
    print(address)
    
    # create URL
    url = f"http://api.openweathermap.org/geo/1.0/direct?q={address}, USA&appid={API_KEY}"
    response = requests.get(url)
    
    # get data
    data = response.json()

    if data:
        location = data[0]
        
        # parse data
        lat = location["lat"]
        lon = location["lon"]
        city = location["name"]
        state = location["state"]
        country = location["country"]
        

        #make row/put into flat dictionary
        row = {"latitude": lat,
              "longitude": lon,
              "city": city,
              "state": state,
              "country": country}
        
        # append to list
        rows.append(row)
        
    else: 
        row = {"latitude": np.nan,
              "longitude": np.nan,
              "city": None,
              "state": None,
              "country": None}
        
        # append to list
        rows.append(row)

    #sleep
    time.sleep(1)
```

```python
loc_df = pd.DataFrame(rows)
loc_df
```
